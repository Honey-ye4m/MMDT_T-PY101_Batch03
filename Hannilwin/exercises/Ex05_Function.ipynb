out an m× n box consisting of asterisks. Shown below is the output of rectangle(2,4)

:Ex 2: Write a function that takes an integer n and returns a random integer with exactly n digits. For
instance, if n is 3, then 125 and 593 would be valid return values, but 093 would not because
that is really 93, which is a two-digit number.

```python
import random

def random_digit(n):
    if n <= 0:
        raise ValueError("n must be a positive integer")

    lower = 10**(n - 1)
    upper = 10**n - 1

    return random.randint(lower, upper)
```

Ex 3: Write a function called number_of_factors that takes an integer and

*   List item
*   List item

returns how many
factors the number has.

```python
def number_of_factors(n):
    if n == 0:
        return 0

    n = abs(n)
    count = 0

    for i in range(1, n + 1):
        if n % i == 0:
            count += 1

    return count
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
def solve_system(a, b, c, d, e, f):
    D = a * d - b * c

    if D == 0:
        return "No unique solution"

    x = (e * d - b * f) / D
    y = (a * f - e * c) / D

    return x, y
```

Ex8:  Write a function called primes that is given a number n and returns a

1.   List item
2.   List item

list of the first n  primes. Let the default value of n be 100.

```python
import math

def is_prime(num):
    if num < 2:
        return False

    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False

    return True


def primes(n=100):
    prime_list = []
    candidate = 2

    while len(prime_list) < n:
        if is_prime(candidate):
            prime_list.append(candidate)
        candidate += 1

    return prime_list
```

Ex9: A Tic-tac-toe board can be represented be a 3×3 two-dimensional list, where zeroes stand for  empty cells, ones stand for X’s and twos stand for O’s. Write a function that is given such a list and randomly chooses a spot in which to place  a 2 (O). The spot chosen must currently be empty and a spot must be chosen.

```python
import random

def place_o(board):
    empty_spots = []

    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                empty_spots.append((i, j))

    # Choose a random empty spot
    row, col = random.choice(empty_spots)

    board[row][col] = 2

    return board
```

Ex10:  Write a function that is given a 9 × 9 potentially solved Sudoku and returns True if it is solved correctly and False if there is a mistake. The Sudoku is correctly solved if there are no repeated numbers in any row or any column or in any of the nine “blocks.”


```python
def is_valid_sudoku(board):
    # Check rows
    for row in board:
        if sorted(row) != list(range(1, 10)):
            return False

    # Check columns
    for col in range(9):
        column = [board[row][col] for row in range(9)]
        if sorted(column) != list(range(1, 10)):
            return False

    # Check 3x3 blocks
    for block_row in range(0, 9, 3):
        for block_col in range(0, 9, 3):
            block = []
            for i in range(3):
                for j in range(3):
                    block.append(board[block_row + i][block_col + j])
            if sorted(block) != list(range(1, 10)):
                return False

    return True
```
