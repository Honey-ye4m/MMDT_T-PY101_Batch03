# Ex 3: Temperature Converter

Write a program that convert temperature in Feranhite to celsius.

```python

fahrenheit = float(input("Enter temperature in Fahrenheit: "))

celsius = (fahrenheit - 32.0) * 5.0 / 9.0

print(f"{fahrenheit}°F is equal to {celsius:.2f}°C")
```

# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
apple_qty= int (input("Enter the Apple Qty"))
apple_price = int ( input("Enter the apple price"))

orange_qty= int (input("Enter the Orange Qty"))
orange_price = int ( input("Enter the orange price"))

mango_qty= int (input("Enter the Mango Qty"))
mango_price = int ( input("Enter the mango price"))

apple_cost = apple_qty * apple_price
orange_cost = orange_qty * orange_price
mango_cost = mango_qty * mango_price

Total_cost = apple_cost + orange_cost + mango_cost

print ("Total_costs =", Total_cost)
```
